```python
%matplotlib widget

import pandas as pd
import datetime
import matplotlib.pyplot as plt
import random
from statistics import mean

import util.benchmark as benchmark
import util.plotting as plotting

import math
```

```python
N = 100000
ITS = 100
Ns = list(range(1,N+1, int(N/ITS)))
```

```python
cases = [([e for e in range(i)], i) for i in Ns]
```

# Busqueda lineal

```python
def findElement(elements, elem):
    for e in elements:
        if e == elem:
            return True
        
    return False
```

```python
linearResults = benchmark.benchmarkFuncForCases(100, findElement, cases)
```

```python
linearCaseFunc = lambda list,_: len(list)
on = "N"

plotting.plotFuncBigOAnalysis(Ns, cases, linearResults, on, linearCaseFunc,int(ITS/2))
```

# Busqueda binaria

```python
def findElement(nums, target):
        l = 0
        r = len(nums)-1
        while l <= r:
            m = int((l+r)/2)
            me = nums[m]
            
            if me == target:
                return True
            if me < target:
                l = m+1
            else:
                r = m-1

        return False
```

```python
binaryResults = benchmark.benchmarkFuncForCases(100, findElement, cases)
```

```python
binaryCaseFunc = lambda list,_: math.log2(len(list))
ologn = "log(N)"

plotting.plotFuncBigOAnalysis(Ns, cases, binaryResults, ologn, binaryCaseFunc,ITS-1)
```

# Comparacion entre ambos

```python
df = pd.DataFrame(list(zip(linearResults, binaryResults)), index = Ns)
ax = df.plot() 
```
