# Naive

```python
def naiveMaxSum(nums, target):
    for i in range(len(nums)):
        for j in range(i+1, len(nums)):
            if nums[i] + nums[j] == target:
                return (i,j)
    return None
```

```python
print(naiveMaxSum([2,7,11,15], 26))
```

## Complejidad

```python
N = 10000
ITS = 30
Ns = list(range(1,N+1, int(N/ITS)))
```

```python
cases = [([e for e in range(n)], 2*n) for idx, n in enumerate(Ns)]
results = benchmark.benchmarkFuncForCases(1, naiveMaxSum, cases, copyCase = False)
```

# Two pointer

```python
def maxSum(nums, target):
    l = 0
    r = len(nums)-1
    while l < r:
        s = nums[l] + nums[r]
        if s == target:
            return (nums[l], nums[r])
        if s < target:
            l+=1
        else:
            r-=1
    return None
```

## Complejidad

```python
N = 1000000
ITS = 30
Ns = list(range(1,N+1, int(N/ITS)))
```

```python
cases = [([e for e in range(n)], 2*n) for idx, n in enumerate(Ns)]
results = benchmark.benchmarkFuncForCases(1, maxSum, cases, copyCase = False)
```

```python
import math
on = lambda elems, _: len(elems)
plotting.plotFuncBigOAnalysis(Ns, cases, results, "N", on, 20)
```
