# Algoritmo de maximo lineal

```python
def maxElement(amount, elems):
    maxElem = elems[0]
    for e in elems:
        maxElem = max(maxElem, e)
        
    return maxElem
```

```python
N = 100000
ITS = 100
Ns = list(range(1,N+1, int(N/ITS)))
```

## Peor caso / Mejor caso / Caso promedio

```python
worstCases = [(i, [e for e in range(i)]) for i in Ns]
worstResults = benchmark.benchmarkFuncForCases(10, maxElement, worstCases)
```
