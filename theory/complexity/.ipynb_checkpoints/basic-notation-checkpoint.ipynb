# Notacion basica de Complejidad
Vamos a explorar como es la notacion basica para el analisis de complejidad de los algoritmos. Para esto veremos los 3 posibles:
- **Big O**: Se utiliza para acotar superiormente el crecimiento de una funcion.
- **Big Ω (Omega)**: Se utiliza para acotar inferiormente el crecimiento de una funcion.
- **Big Θ (Theta)**: Se utiliza para acotar tanto superior como inferiormente el crecimiento de una funcion.

Veremos estas notaciones con un algoritmo sencillo, donde la funcion que se menciono previamente va a representar el tiempo que toma nuestro algoritmo para computar un resultado.

```python
%matplotlib widget

import pandas as pd
import datetime
import matplotlib.pyplot as plt
import random
from statistics import mean

import util.benchmark as benchmark
import util.plotting as plotting
```

# Algoritmo de busqueda lineal
Tenemos la siguiente implementacion para decidir dada una lista y un elemento, si este elemento pertenece a la lista.

```python
def findElement(elements, elem):
    for e in elements:
        if e == elem:
            return True
        
    return False
```

## Datos de entrada
Tenemos listas de distintas longitudes, y un elemento a encontrar.
- Cada una de estas listas va a tener una longitud, llamemosla **N**.
- El elemento a buscar siempre va a ser 1.

```python
N = 100000
ITS = 100
Ns = list(range(1,N+1, int(N/ITS)))
```

## Caso peor, mejor y promedio
Cuando pensamos en un algoritmo, este normalmente tiene distintas familias de datos de entrada. En cada uno de estos, el algoritmo puede comportarse de manera completamente distinta, en terminos de eficiencia.
Veamos cada uno de estos casos.

### Peor caso: No existe el elemento / esta al final
En estas entradas, el elemento que se busca o bien no existe, o si pero se encuentra al final. En cualquiera de estos tenemos que iterar toda la lista.

```python
worstCases = [([e for e in range(i)], i) for i in Ns]
worstResults = benchmark.benchmarkFuncForCases(100, findElement, worstCases)
```

```python
worstCaseFunc = lambda list,_: len(list)
worstCaseFuncSymbol = "N"

plotting.plotFuncComplexityAnalysis(Ns, worstCases, worstResults, 
                                   worstCaseFuncSymbol, worstCaseFunc,
                                   worstCaseFuncSymbol, worstCaseFunc,
                                   worstCaseFuncSymbol, worstCaseFunc, int(ITS/2))
```

### Mejor caso: Es el primer elemento
En estas entradas, el elemento que se busca es el primero, por lo que terminamos la busqueda instantaneamente.

```python
bestCases = [([e for e in range(i)], 0) for i in Ns]
bestResults = benchmark.benchmarkFuncForCases(1000, findElement, bestCases)
```

```python
bestCaseFunc = lambda list,_: 1
bestCaseSymbol = "1"

plotting.plotFuncComplexityAnalysis(Ns, bestCases, bestResults, 
                                   bestCaseSymbol, bestCaseFunc,
                                   bestCaseSymbol, bestCaseFunc,
                                   bestCaseSymbol, bestCaseFunc,
                                   int(ITS/2))
```

### Caso promedio: Esta distribuido uniformemente en lista
Si asumimos que el elemento que nos piden pertenece, y la distribucion de los numeros de la lista responde a una distribucion uniforme, entonces va a tender a recorre la mitad de la lista en promedio.

```python
avgCases = [([e for e in range(i)], 0) for i in Ns]
avgResultsRuns = []
for _ in range(20):
    avgCasesRun = [([e for e in range(i)], random.randint(0,i)) for i in Ns]
    avgResultsRun = benchmark.benchmarkFuncForCases(10, findElement, avgCasesRun)
    
    avgResultsRuns.append(avgResultsRun)
avgResults = list(map(mean, zip(*avgResultsRuns)))
```

```python
on = lambda list,_: len(list)

plotting.plotFuncComplexityAnalysis(Ns, avgCases, avgResults, "N", on, "N", on, "N", on, 90)
```
