```python
%matplotlib widget
# %matplotlib ipympl

import pandas as pd
import datetime
import matplotlib.pyplot as plt
import random
from statistics import mean

import util.benchmark as benchmark
import util.plotting as plotting

import math
from asyncio import PriorityQueue
```

# Priority Queue

```python
N = 500000
ITS = 100
Ns = list(range(1,N+1, int(N/ITS)))
```

## Insercion

```python
def insert(elem, pq):
    pq.put_nowait(elem)
```

```python
insertCases = []
for i in Ns:
    pq = PriorityQueue()
    for e in range(i):
        pq.put_nowait(e)
    insertCases.append((i, pq))
insertResults = benchmark.benchmarkFuncForCases(1, insert, insertCases, copyCase = True)
```

```python
ologn = lambda _, pq: math.log2(pq.qsize())

plotting.plotFuncBigOAnalysis(Ns, insertCases, insertResults, "log(N)", ologn, 1)
```

## Eliminacion

```python
def remove(elem, pq):
    pq.get_nowait()
```

```python
removeCases = []
for i in Ns:
    pq = PriorityQueue()
    for e in range(i):
        pq.put_nowait(e)
    removeCases.append((i, pq))
removeResults = benchmark.benchmarkFuncForCases(1, remove, removeCases, copyCase = True)
```

```python
ologn = lambda _, pq: math.log2(pq.qsize())

plotting.plotFuncBigOAnalysis(Ns, removeCases, removeResults, "log(N)", ologn, 1)
```
