# Dictionary

## Insercion

```python
def insert(e, d):
    d[e] = e
```

```python
N = 100000
ITS = 100
Ns = list(range(1,N+1, int(N/ITS)))
insertCases = [(i-1, dict([(e,e) for e in range(i)])) for i in Ns]
insertResults = benchmark.benchmarkFuncForCases(1, insert, insertCases, copyCase = True)
```

```python
o1 = lambda e, d: 1

plotting.plotFuncBigOAnalysis(Ns, insertCases, insertResults, "1", o1, 1)
```

## Busqueda
