```python
%matplotlib widget
# %matplotlib ipympl

import pandas as pd
import datetime
import matplotlib.pyplot as plt
import random
from statistics import mean

import util.benchmark as benchmark
import util.plotting as plotting
```

```python
N = 1000
ITS = 100
Ns = list(range(1,N+1, int(N/ITS)))
```

# Selection Sort


```python
def selectionSort(elems, compareFunc):
    for i in range(len(elems)):
        bestIdx = i
        for j in range(i, len(elems)):
            if compareFunc(elems[j], elems[bestIdx]):
                bestIdx = j

        elems[bestIdx], elems[i] = elems[i], elems[bestIdx]

    return elems
```

## Peor caso / Mejor caso

```python
comparisonsPerCase = [0 for _ in Ns]
def getCompareFunc(i):
    def compareFunc(a,b):
        comparisonsPerCase[i]+=1
        return a < b

    return compareFunc
    
cases = [([e for e in range(n)], getCompareFunc(idx)) for idx, n in enumerate(Ns)]
results = benchmark.benchmarkFuncForCases(1, selectionSort, cases, copyCase = True)
comparisonsPerCase = list(map(lambda c: c / 10, comparisonsPerCase))
```

```python
on2 = lambda elems, _: len(elems)**2

plotting.plotFuncBigOAnalysis(Ns, cases, results, "N²", on2, 25)
```

### Con respecto a las comparaciones

```python
plotting.plotFuncBigOAnalysis(Ns, cases, comparisonsPerCase, "N²", on2, 25)
```

# Insertion Sort

```python
def insertionSort(elems, compareFunc):
    for i in range(len(elems)):
        curIdx = i
        while curIdx > 0 and compareFunc(elems[curIdx], elems[curIdx-1]):
            elems[curIdx-1], elems[curIdx] = elems[curIdx], elems[curIdx-1]
            curIdx-=1

    return elems
```

## Mejor caso

```python
comparisonsPerCase = [0 for _ in Ns]
def getCompareFunc(i):
    def compareFunc(a,b):
        comparisonsPerCase[i]+=1
        return a < b

    return compareFunc
    
cases = [([e for e in range(n)], getCompareFunc(idx)) for idx, n in enumerate(Ns)]
results = benchmark.benchmarkFuncForCases(10, insertionSort, cases, copyCase = True)
comparisonsPerCase = list(map(lambda c: c / 10, comparisonsPerCase))
```

```python
on = lambda elems, _: len(elems)

plotting.plotFuncBigOAnalysis(Ns, cases, results, "N", on, 50)
```

### Con respecto a las comparaciones

```python
plotting.plotFuncBigOAnalysis(Ns, cases, comparisonsPerCase, "N", on, 25)
```

```python
comparisonsPerCase = [0 for _ in Ns]
def getCompareFunc(i):
    def compareFunc(a,b):
        comparisonsPerCase[i]+=1
        return a < b

    return compareFunc
    
cases = [([n-e for e in range(n)], getCompareFunc(idx)) for idx, n in enumerate(Ns)]
results = benchmark.benchmarkFuncForCases(1, insertionSort, cases, copyCase = True)
comparisonsPerCase = list(map(lambda c: c / 10, comparisonsPerCase))
```
