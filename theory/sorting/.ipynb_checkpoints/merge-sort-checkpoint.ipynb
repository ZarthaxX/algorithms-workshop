# Merge Sort

```python
someGraph = statesgraph.StatesGraph()
@statesplotting.trackStatesGraph(someGraph)
def mergeSort(elems):
    if len(elems) < 2:
        return elems

    mid = int(len(elems)/2)
    leftHalf = mergeSort(elems[:mid])
    rightHalf = mergeSort(elems[mid:])
    mergedElems = []
    while len(leftHalf) > 0 and len(rightHalf) > 0:
        if leftHalf[0] < rightHalf[0]:
            mergedElems.append(leftHalf[0])
            leftHalf.pop(0)
        else:
            mergedElems.append(rightHalf[0])
            rightHalf.pop(0)
    
    if len(leftHalf) > 0:
        mergedElems += leftHalf
    if len(rightHalf) > 0:
        mergedElems += rightHalf
    return mergedElems
```

```python
print(mergeSort([1,6,4,3,8,2,7,5]))
net = statesplotting.drawTreeGraph(someGraph)
net.show("mergesort.html")
```

```python
def mergeSort(elems, compareFunc):
    if len(elems) < 2:
        return elems

    mid = int(len(elems)/2)
    leftHalf = mergeSort(elems[:mid],compareFunc)
    rightHalf = mergeSort(elems[mid:],compareFunc)
    mergedElems = []
    while len(leftHalf) > 0 and len(rightHalf) > 0:
        if compareFunc(leftHalf[0], rightHalf[0]):
            mergedElems.append(leftHalf[0])
            leftHalf.pop(0)
        else:
            mergedElems.append(rightHalf[0])
            rightHalf.pop(0)
    
    if len(leftHalf) > 0:
        mergedElems += leftHalf
    if len(rightHalf) > 0:
        mergedElems += rightHalf
    return mergedElems
```

## Peor caso / Mejor caso

```python
comparisonsPerCase = [0 for _ in Ns]
def getCompareFunc(i):
    def compareFunc(a,b):
        comparisonsPerCase[i]+=1
        return a < b

    return compareFunc
    
cases = [([e for e in range(n)], getCompareFunc(idx)) for idx, n in enumerate(Ns)]
results = benchmark.benchmarkFuncForCases(1, mergeSort, cases, copyCase = True)
comparisonsPerCase = list(map(lambda c: c / 10, comparisonsPerCase))
```

```python
import math
onlogn = lambda elems, _: len(elems) * math.log2(len(elems))
plotting.plotFuncBigOAnalysis(Ns, cases, results, "Nlog(N)", onlogn, 20)
```
